```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset["CKD_yes"].value_counts()
```

```python
indep=dataset[["age","bp","al","su","bgr","bu","sc","sod","pot","hrmo","pc_normal","pcc_present","ba_present","htn_yes","dm_yes","cad_yes","appet_yes","pe_yes","ane_yes"]]
dep=dataset["CKD_yes"]
```

```python
indep.shape
```

```python
dep
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(indep, dep, test_size = 1/3, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'criterion': ['gini', 'entropy'],  'max_features': ['auto', 'sqrt', 'log2'],'n_estimators': [10, 100]}
grid = GridSearchCV(RandomForestClassifier(random_state=0), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(indep, dep)
```

```python
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),param_grid=param_grid,verbose=2,n_jobs=-1)
```

```python
y_pred = grid.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_test, y_pred)
```

```python
print(clf_report)
```
